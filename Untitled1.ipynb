```python
import dask.dataframe as dd
from dask import persist
from collections import defaultdict

ericsson = dd.read_parquet('data.parquet')
ericsson.persist()
```

```python
teleMetry = ericsson.telemetry_log
```

```python
teleMetry.compute()
```

```python
%%time
tele_dict = defaultdict(lambda : 0)
for dictionary in teleMetry:
    for name, value in dictionary.items():
        if value is not None:
            tele_dict[name] += 1
tele_dict
```

```python
%%time
tele_dict2 = {}
for dictionary in teleMetry:
    for name, value in dictionary.items():
        if value is not None:
            if name not in tele_dict2:
                tele_dict2[name] = 1
            else:
                tele_dict2[name] += 1
tele_dict2
```

```python
%%time 
from collections import Counter

counter = Counter()
for dictionary in teleMetry:
    counter.update([name for name,value in dictionary.items() if value is not None])
            
counter
```
